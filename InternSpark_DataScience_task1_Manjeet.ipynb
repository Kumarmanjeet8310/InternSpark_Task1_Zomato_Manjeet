```python
import pandas as pd
df=pd.read_csv("zomato.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
print(df.shape)
```

```python
df.shape
```

```python
df=df.dropna()
print(df.shape)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
df['rest_type'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Restaurant Types")
plt.xlabel("Restaurant Types")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,5))
df['location'].value_counts().head(10).plot(kind='bar',color='orange')
plt.title("Top 10 Restaurant Locations")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['rate'].value_counts().head(10).plot(kind='bar',color='green')

plt.title("Top Restaurants Ratings")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,5))
df['cuisines'].value_counts().head(10).plot(kind='bar',color='purple')
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['approx_cost(for two people)'].value_counts().head(10).plot(kind='bar',color='red')
plt.title("Approx Cost for Two People")
plt.xlabel("Cost")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
